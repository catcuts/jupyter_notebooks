<article class="post-block"><h1 class="post-title">Mosquitto MQTT 安裝</h1><span class="post-info">2016年8月17日</span><div class="post-content"><h6> <i class="fa fa-tags">Tags :<code>Ubuntu</code><code>Mosquitto</code><code>MQTT</code></i></h6><ul>
<li><p>MQTT 的全名為 Message Queuing Telemetry Transport，為 IBM 和 Eurotech 共同製定出來的協定，MQTT 是為了物聯網設計的協定，透過Publish/Subscribe 的方式來做訊息傳送。</p>
</li>
<li><p>Publish/Subscribe 有三種主要的組成元件，分別為Publisher、Subscriber 以及 Topic。</p>
</li>
<li><p><a href="https://mosquitto.org/" target="_blank" rel="external">mosquitto</a> 為 Eclipse 所開發的開源 MQTT broker 和給用戶端使用C語言的API介面，目前最新版本為 V1.4.9。</p>
</li>
</ul>

<p>Broker的認證方式，Mosquitto端提供了四種認證方法分別為</p>
<ul>
<li>無認證</li>
<li>帳號/密碼</li>
<li>預先共享密鑰(Pre-shared Key)加密</li>
<li>憑證加密</li>
</ul>
<h2 id="無認證"><a href="#無認證" class="headerlink" title="無認證"></a>無認證</h2><p>要打開 <code>/etc/mosquitto/mosquitto.conf</code> 設定檔將allow_anonymous設定為true即可。此時我們可以利用mosquitto_sub來測試   </p>
<p>Subscribe<br></p><figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_sub -h <span class="string">"&lt;IP address&gt;"</span> -v -t <span class="string">"&lt;Topic's name&gt;"</span></div></pre></td></tr></tbody></table></figure><p></p>
<p>Publish<br></p><figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_pub -h <span class="string">"&lt;IP address&gt;"</span> -v -t <span class="string">"&lt;Topic's name&gt;"</span> -m <span class="string">"&lt;Message&gt;"</span></div></pre></td></tr></tbody></table></figure><p></p>
<p><code>-t :訂閱的主題，-v :連主題名稱也列印在銀幕上。</code></p>
<h2 id="帳號-密碼"><a href="#帳號-密碼" class="headerlink" title="帳號/密碼"></a>帳號/密碼</h2><ul>
<li><p>新增帳號密碼</p>
<p>  使用此方式需利用上述的 <code>mosquitto_passwd</code> 產生密碼，並將密碼檔案名稱設定到password_file此欄位。</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_passwd -c <span class="string">"&lt;passwordfile&gt;"</span> <span class="string">"&lt;username&gt;"</span></div></pre></td></tr></tbody></table></figure>
<blockquote>
<p>如果passwordfile檔案已經存在，則下次新增使用者則不需要加 -c</p>
</blockquote>
<p>  下列命令將新產生一個密碼檔案/etc/mosquitto/passwd，並建立一個帳號名稱為easy，使用者輸入密碼後即可建立帳號與密碼</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div><div class="line">2</div><div class="line">3</div></pre></td><td class="code"><pre><div class="line">sudo mosquitto_passwd -c /etc/mosquitto/passwd easy</div><div class="line">Password: xxxx</div><div class="line">Reenter password: xxxx</div></pre></td></tr></tbody></table></figure>
</li>
<li><p>設定ACL(Access Control List)</p>
<p>  Path : <code>/etc/mosquitto/acl</code> 內容如下:</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div><div class="line">2</div><div class="line">3</div><div class="line">4</div><div class="line">5</div><div class="line">6</div><div class="line">7</div></pre></td><td class="code"><pre><div class="line"><span class="comment"># 用戶easy 對於主題hello/world 能夠進行寫入</span></div><div class="line">user easy  </div><div class="line">topic write hello/world</div><div class="line"></div><div class="line"><span class="comment"># 用戶easy 對於主題hello/world 能夠進行讀取</span></div><div class="line">user easy  </div><div class="line">topic <span class="built_in">read</span> hello/world</div></pre></td></tr></tbody></table></figure>
</li>
<li><p>最後更改 <code>/etc/mosquitto/mosquitto.conf</code> 設定檔</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div><div class="line">2</div><div class="line">3</div><div class="line">4</div><div class="line">5</div></pre></td><td class="code"><pre><div class="line">allow_anonymous <span class="literal">false</span>  <span class="comment">#不允許匿名傳輸</span></div><div class="line"></div><div class="line">password_file /etc/mosquitto/passwd	<span class="comment">#配置用戶(username)密碼</span></div><div class="line"></div><div class="line">acl_file /etc/mosquitto/acl	<span class="comment">#設置用戶與主題(topic)存取權限</span></div></pre></td></tr></tbody></table></figure>
</li>
<li><p>測試</p>
<p>  Subscribe</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_sub -h <span class="string">"&lt;IP address&gt;"</span> -v <span class="_">-d</span> -t <span class="string">"&lt;topic's name'&gt;"</span> -u <span class="string">"&lt;User&gt;"</span> -P <span class="string">"&lt;Possword&gt;"</span></div></pre></td></tr></tbody></table></figure>
<p>  Publish</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_pub -h <span class="string">"&lt;IP address&gt;"</span> -v -t <span class="string">"&lt;Topic's name&gt;"</span> -m <span class="string">"&lt;Message&gt;"</span> -u <span class="string">"&lt;User&gt;"</span> -P <span class="string">"&lt;Possword&gt;"</span></div></pre></td></tr></tbody></table></figure>

<h2 id="預先共享密鑰-Pre-shared-Key-加密"><a href="#預先共享密鑰-Pre-shared-Key-加密" class="headerlink" title="預先共享密鑰(Pre-shared Key)加密"></a>預先共享密鑰(Pre-shared Key)加密</h2><p>這個認證方式需要使用到OpenSSL中的預先共享密鑰加密方式，而在OpenBSD開發的LibreSSL專案中，因為SSL/TLS安全性的問題，預先共享密鑰加密方式暫時被移除了。</p>

<h2 id="憑證加密"><a href="#憑證加密" class="headerlink" title="憑證加密"></a>憑證加密</h2><p>TLS是非常成熟的安全協議，在握手的時候便可以創建安全連接，使得駭客無法竊聽或者篡改內容。使用TLS的時候有以下注意點：</p>
<ol>
<li>盡可能使用高版本的TLS。</li>
<li>驗證X509憑證連結防止中間人攻擊。</li>
<li>儘量使用有CA發佈的證書。</li>
</ol>
<p>產生憑證的方法可以參考如下</p>
<ul>
<li><p>產生憑證</p>
<p>  建立CA(Certificate Authority)所需的證書及密鑰</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">openssl req -new -x509 -days 365 -extensions v3_ca -keyout ca.key -out ca.crt</div></pre></td></tr></tbody></table></figure>
<p>  系統端的憑證</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div><div class="line">2</div><div class="line">3</div><div class="line">4</div><div class="line">5</div><div class="line">6</div><div class="line">7</div><div class="line">8</div></pre></td><td class="code"><pre><div class="line"><span class="comment"># 產生不加密的金鑰</span></div><div class="line">openssl genrsa -out server.key 2048</div><div class="line"></div><div class="line"><span class="comment"># 產生憑證簽發請求檔案(Certificate Signing Request)</span></div><div class="line">openssl req -new -key server.key -out server.csr</div><div class="line"></div><div class="line"><span class="comment"># 憑證簽發</span></div><div class="line">openssl x509 -req -in server.csr -CA ca.crt -CAkey ca.key -CAcreateserial -out server.crt -days 365</div></pre></td></tr></tbody></table></figure>
</li>
<li><p>修改 <code>/etc/mosquitto/mosquitto.conf</code> 設定檔</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div><div class="line">2</div><div class="line">3</div><div class="line">4</div><div class="line">5</div><div class="line">6</div><div class="line">7</div></pre></td><td class="code"><pre><div class="line">listener 8883</div><div class="line"></div><div class="line">cafile /etc/mosquitto/ca.crt</div><div class="line"></div><div class="line">certfile /etc/mosquitto/server.crt</div><div class="line"></div><div class="line">keyfile /etc/mosquitto/server.key</div></pre></td></tr></tbody></table></figure>
</li>
<li><p>測試</p>
<p>  Subscribe</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_sub <span class="_">-d</span> -h <span class="string">"&lt;IP address&gt;"</span> -p 8883 -t topic --cafile ca.crt</div></pre></td></tr></tbody></table></figure>
<p>  Publish</p>
  <figure class="highlight bash"><table><tbody><tr><td class="gutter"><pre><div class="line">1</div></pre></td><td class="code"><pre><div class="line">mosquitto_pub <span class="_">-d</span> -h <span class="string">"&lt;IP address&gt;"</span> -p 8883 -t topic -m “Hello World!” --cafile ca.crt</div></pre></td></tr></tbody></table></figure>
</li>
</ul>
