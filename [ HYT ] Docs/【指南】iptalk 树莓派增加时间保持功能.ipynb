### <span style='color: #dd4b39;'>2. 更新安装包</span>

1. 插入 U 盘，然后 putty 进入树莓派系统
<br><br>
2. 把根目录重载为可写：`mount -o remount,rw /`
<br><br>
3. 创建 U 盘挂载点：`if [ ! -d /home/pi/usb ]; then mkdir -p /home/pi/usb; fi`
<br><br>
4. 找到 U 盘设备名称：`fdisk -l | grep ^Disk`

    注：
    - 有可能是 `/dev/sda` 也有可能是 `/dev/sdb` 注意看
    - 以下以 `/dev/sda` 为例
<br><br>
4. 找到 U 盘分区：`fdisk -l`

    注：
    - 有可能是 `/dev/sda` 或 `/dev/sda1` 也有可能是 `/dev/sda2` 注意看
    - 以下以 `/dev/sda` 为例
<br><br>
5. 挂载所选 U 盘：`mount /dev/sda /home/pi/usb`
<br><br>
6. 进入 U 盘：`cd /home/pi/usb`
<br><br>
7. 运行：`bash inject_new_maker.sh`
<br><br>
8. 退出 U 盘：`cd ~`
<br><br>
9. 弹出 U 盘：`umount /home/pi/usb`
<br><br>
10. 把根目录重载为只读：`mount -o remount,ro /`

### <span style='color: #dd4b39;'>3. 配置系统</span>

1. 执行 `mount -o remount,rw /`
<br><br>
2. 执行 `mount -o remount,rw /boot`
<br><br>
3. 执行 `raspi-config`
<br><br>
4. 选择 `5 Interfacing Options` 回车
<br><br>
5. 选择 `P5 I2C` 回车
<br><br>
6. 选择 `Yes`
<br><br>
7. 选择 `OK`
<br><br>
8. 选择 `finish`

### <span style='color: #dd4b39;'>5. 最小测试流程</span>

1. ssh 进入树莓派，修改系统时间为错误的时间：`sudo date --s="2018-07-03 11:18:00"`
<br><br>
2. web 管理界面 > 系统配置 > 系统设置 > (拉到底) > 校正时间
<br><br>
3. ssh 进入树莓派，执行 `date` 检查系统时间与 web 管理界面所在电脑的时间一致
 - 如果不一致，执行 `hwclock -r -D` 查看原因
<br><br>
4. 执行 ```if [ "`mount | grep ^/dev/mm | grep rw`" == "" ]; then echo ok; fi``` 输出 `ok`
<br><br>
5. 断电一段时间后，ssh 进入树莓派，执行 `date` 检查系统时间与 web 管理界面所在电脑时间一致
 - 如果不一致，执行 `hwclock -r -D` 查看原因
