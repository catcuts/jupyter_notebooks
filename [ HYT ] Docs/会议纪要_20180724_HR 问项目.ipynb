**hr**：**<span style="font-size: large;">负责项目概述？</span>**

**me**：
1. 云平台的解释

    1）云平台项目是为了交付一个用于接入各种物联网设备而设立的项目。

    2）物联网设备可以是主机、分机、对讲设备、视频监控等。

    3）当前云平台的开发主要围绕医疗项目进行。
<br><br>
2. 云平台的开发分为四块：

    1）后端服务器开发（目前转为 NodeJs 语言，之前为 Python）
    
    2）前端页面开发（JavaScript 语言）
    
    3）主机开发（Android 语言）
    
    4）分机开发（嵌入式语言，如 C++）
