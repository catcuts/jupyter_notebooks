### 详

#### 1. **病床分机测试反馈的 10 个 bug 中优先级较高的 5 个**

- 分机拔掉网线后，重新插上，mqtt服务没有连接上

- 分机在其他页面浏览时，按面板呼叫键，浏览页与呼叫页面重叠

- 病床分机页面处于校准页面，短按复位键，病床分机退出触摸校准 进入病床分机页面时，经常按了一下之后就会卡住，接着按没反应，或出现错位

- 病床分机注册成功时，面板指示灯，病床分机面板的按键指示灯不会显示红色慢闪  								
- 病床分机呼叫医护主机，面板指示灯状态不对
