## 二、智能网关的运行环境部署

　　**时间**：6.14 ～ 6.18

　　**内容**：

　　　　智能网关需要运行在 OPi 开发板上，需要的环境有：

　　　　　　1. Nodejs 10 以上（难点：系统自带包管理器安装无能，只可手动安装）

　　　　　　2. GPIO 通信环境（难点：寻找可用的第三方库不易，最终选择 pyA20 并测试成功）

　　　　　　3. 串口通信环境（难点：系统未默认开启可供开发的串口且未提示，多方尝试后开启成功并测试成功）

　　　　　　4. 正常联网环境（难点：联网设置与其它发行版有所差异，尝试多种方法后成功联网）

　　**难点**：

　　　　1. 上已述。
