## 回顾

　　上半年我们云平台开发组的工作可以分为两条线，一条是主线，另一条是辅线。

　　云平台2.0开发为主线，TB8000——也是云平台1.0的一部分——的维护为辅线。

　　云平台2.0是我们总结了1.0的经验之后，带来的升级版本。它采取全新的语言、优化的架构以及前瞻的理念，旨在打造一个功能丰富、容易扩展、方便维护、接口友好、并且不限接入目标的云平台。它的高要求，决定了它大工作量。从去年11月21号项目创建之日起，我们已经奋斗了122个有效天。目前，它尚未偏离我们的初衷，并且在大量基础工作的支持下，已经能提供一些初步的服务，如医疗设备的接入服务，以及为医疗项目提供 http 数据请求等服务。**接下来，我们还将进一步加大联调的深度，并根据联调的反馈和产品经理提出的需求，继续云平台的业务开发，同时促进完善云平台的功能。同时，我们还会在原有前端管理页面的基础上，继续开发 Web 端，以便客户对设备进行监控。再接下来，我们还会进一步提升云平台的性能，比如提高通信安全性、加强权限控制、引入分布式和高并发、并实现合理负载均衡等，无限接近我们的既定目标。**

　　我希望与你分享，上半年这122个有效天我们做了什么。为此，我整理了一个列表：

**回 顾**

- 2017-11-21：项目创建；
- 2017-11-23：（2天后）测试框架出现；

- 2018-06-10：（20天后）医疗设备可以通过 http 访问数据库数据。

　　可以看到，到目前为止，我们用了 70% 的时间来打造地基，而地基越深，高楼就越稳。在云平台持续开发的过程中，接入的应用和设备将越来越多，有朝一日，我们的云平台势必高楼林立，彼时，将是一副万物互联、生机勃勃的景象。

　　反思部分主要是关于上半年的工作辅线，也就是 TB8000 的维护工作。

　　我相信，当提到“树莓派”一词时，有不少人会如同谈狐色变一般，避犹不及。

　　这是我的两方面错误造成的。

　　首先，是对一些在测试过程中出现的现象，没有充分细究，以为没有复现就算了，结果导致出厂后大面积出现问题。

　　然后，是在传达技术内容时表达欠佳，以致于相关的技术培训和交流都收效甚微，导致有些支持工作不得不转移给自己来做，这直接占据了主线工作的时间。

　　此次反思之后，在今后的工作过程中，不放过产品现象的蛛丝马迹将成为我的态度，时刻不忘提高表达能力将成为我的习惯。

　　贯彻实施上述态度和习惯的前提，无疑是超高的工作效率，毕竟一天只有 24 小时。所以提升工作效率将成为亟待提升的技能。

　　为此，加油。
