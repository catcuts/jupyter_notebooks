### <span style='color: #dd4b39;'>2. 更新安装包</span>

1. putty 登入，把根目录重载为可写：`mount -o remount,rw /`
<br><br>
2. filezilla 登入，进入文件夹 `/home/pi/`
<br><br>
3. 再进入文件夹 `make_ro_system_for_rpi3_jessie`（如果不存在则创建）
<br><br>
4. `ctrl + A` 选中 该文件夹内 所有文件，删除之（确认好再删除，如果没有文件则无需删除）
<br><br>
5. 把 `make_ro_system_for_rpi3_jessie.zip` 上传到 该文件夹内
<br><br>
6. putty 进入该文件夹：`cd /home/pi/make_ro_system_for_rpi3_jessie`
<br><br>
7. 解压缩：`unzip make_ro_system_for_rpi3_jessie.zip`
<br><br>
8. 把根目录重载为只读：`mount -o remount,ro /`

___
___

### <span style='color: #dd4b39;'>3. 配置系统</span>

1. 执行 `mount -o remount,rw /`
<br><br>
2. 执行 `mount -o remount,rw /boot`
<br><br>
3. 执行 `raspi-config`
<br><br>
4. 选择 `5 Interfacing Options` 回车
<br><br>
5. 选择 `P5 I2C` 回车
<br><br>
6. 选择 `Yes`
<br><br>
7. 选择 `OK`
<br><br>
8. 选择 `Finish`

### <span style='color: #dd4b39;'>5. 最小测试流程</span>

1. ssh 进入树莓派，修改系统时间为错误的时间：`sudo date --s="2018-07-03 11:18:00"`
<br><br>
2. web 管理界面 > 系统配置 > 系统设置 > (拉到底) > 校正时间
<br><br>
3. ssh 进入树莓派，执行 `date` 检查系统时间与 web 管理界面所在电脑的时间一致
 - 如果不一致，执行 `hwclock -r -D` 查看原因
<br><br>
4. 执行 ```if [ "`mount | grep ^/dev/mm | grep rw`" == "" ]; then echo ok; fi``` 输出 `ok`
<br><br>
5. 断电一段时间后，ssh 进入树莓派，执行 `date` 检查系统时间与 web 管理界面所在电脑时间一致
 - 如果不一致，执行 `hwclock -r -D` 查看原因
